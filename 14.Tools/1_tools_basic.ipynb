## Custom Tools

```python
from langchain_core.tools import tool
# step 1 create a function

def sum(a,b):
    """Add two numbers"""
    return a+b
```

```python
# step 2 add type hints

def sum(a:int,b:int) -> int:
    """Add two numbers"""
    return a+b
```

```python
# step 3 add tool decorator
@tool
def sum(a:int,b:int) -> int:
    """Add two numbers"""
    return a+b
```

```python
result = sum.invoke({"a":4,"b":10})
print(result)
```

```python
print(sum.name)
print(sum.description)
print(sum.args)
```

```python
print(sum.args_schema.model_json_schema())
```

## Method 2 - Using StructuredTool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class SumInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
def sum_fun(a:int,b:int)->int:
    return a+b
```

```python
sum_tool = StructuredTool.from_function(
    func=sum_fun,
    name="Addition",
    description="Add two numbers",
    args_schema=SumInput
)
```

```python
result = sum_tool.invoke({"a":5,"b":6})
print(result)
```

```python
print(sum_tool.name)
print(sum_tool.description)
print(sum_tool.args)
```

## Method 3 - Using BaseTool Class

```python
from langchain.tools import BaseTool
from typing import Type
```

```python
# arg schema using pydantic

class SumInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
class SumTool(BaseTool):
    name:str = "Addition"
    description : str ="Add two numbers"
    args_schema: Type[BaseModel] = SumInput
    
    
    def _run(self,a:int,b:int)->int:
        return a+b
    
```

```python
sum_tool = SumTool()
result = sum_tool.invoke({"a":4,"b":6})
print(result)
```

```python
print(sum_tool.name)
print(sum_tool.description)
print(sum_tool.args)
```

## Toolkit

```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add,multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name,"=>",tool.description)
```
